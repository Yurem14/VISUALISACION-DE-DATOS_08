2. Crea un grafico de barras en donde se muestren la cantidad de personas registradas por `occupation`.

```python
In['occupation'].value_counts()
```

```python
a=In['education_level'].value_counts()
plt.figure(figsize=(10,6))  
plt.bar(a.index,a.values,color='red')
plt.title('Profesiones de los clientes')
plt.xlabel('Profeciones')
plt.ylabel('Número de Clientes')
plt.xticks(rotation=90)
plt.show()
```

4. Crea un grafico circular en donde se muestre la proporcion de Hombres y mujeres de acuerdo a la columna `gender`

```python
a=In['gender'].value_counts()
plt.figure(figsize=(10,6))  
plt.pie(a.values, labels = a.index, colors=['red', 'yellow'], autopct='%1.1f%%')
plt.title('Proporcion de Generos')
plt.show()
```
